```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import joblib

X = np.array([[1, 200], [3, 400], [0, 100], [5, 1000], [0, 50]])
y = [1, 1, 0, 1, 0]

clf = RandomForestClassifier()
clf.fit(X, y)
joblib.dump(clf, "ml_model.pkl")
```

```python
# Feature Extraction
def extract_features(content):
    keywords = ["malware", "trojan", "attack", "exploit"]
    score = sum(content.lower().count(k) for k in keywords)
    return {
        "keyword_count": score,
        "length": len(content),
        "suspicious": score > 0
    }

# Analyze Files
def analyze_files(file_paths):
    results = []
    for file_path in file_paths:
        with open(file_path, 'r', errors='ignore') as f:
            content = f.read()
        features = extract_features(content)
        results.append({
            "file": file_path,
            "suspicious": features["suspicious"],
            "keyword_count": features["keyword_count"],
            "length": features["length"],
            "content_sample": content[:200]
        })
    return results

# ML Detection
import numpy as np
import joblib
import os

def ml_anomaly_detection(data):
    if os.path.exists("ml_model.pkl"):
        model = joblib.load("ml_model.pkl")
    else:
        model = None

    for entry in data:
        features = np.array([[entry["keyword_count"], entry["length"]]])
        score = model.predict_proba(features)[0][1] if model else 0.5
        entry["anomaly_score"] = round(score, 2)
        entry["recommendation"] = (
            "Investigate Immediately" if score > 0.7 else
            "Review" if score > 0.4 else
            "Likely Safe"
        )
    return data

# Report Generation
import json
import csv
from fpdf import FPDF

def generate_json(data):
    with open("report.json", "w") as f: json.dump(data, f, indent=4)

def generate_csv(data):
    with open("report.csv", "w", newline='') as f:
        writer = csv.DictWriter(f, fieldnames=data[0].keys())
        writer.writeheader()
        writer.writerows(data)

def generate_pdf(data):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)
    pdf.cell(200, 10, txt="Cyber Triage Report", ln=True, align='C')
    for entry in data:
        for k, v in entry.items():
            pdf.multi_cell(0, 10, f"{k}: {v}")
        pdf.ln()
    pdf.output("report.pdf")

# Full pipeline
results = analyze_files(uploaded_files)
results = ml_anomaly_detection(results)
generate_json(results)
generate_csv(results)
generate_pdf(results)

# Display results
import pandas as pd
pd.DataFrame(results)
```
